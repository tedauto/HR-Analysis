```python
# correlation
corr_matrix = df.corr()
plt.figure(figsize=(20,10))
sns.heatmap(corr_matrix,annot=True,cmap='coolwarm')
plt.title('Correlation map for numeric variables')
plt.show()
```

Multivariate Analysis

```python
# Overtime vs Age
sns.boxplot(data=df,x='OverTime',y = 'Age')
plt.title('Relationship between OverTime and Age')
plt.show()
```

```python
# Income by job role
sns.boxplot(data=df,x='JobRole',y='MonthlyIncome')
plt.title('Income by Job role')
plt.xticks(rotation =90)
plt.show()
```

```python
# Age distribution in the departments
sns.boxplot(data = df, x= 'Department',y ='Age')
plt.title('Age distribution in departments')
plt.show()
```

```python
# attrition by job role
sns.countplot(data=df,x='JobRole',hue ='Attrition')
plt.title('Attrition by employee')
plt.xticks(rotation = 90)
plt.show()
```
